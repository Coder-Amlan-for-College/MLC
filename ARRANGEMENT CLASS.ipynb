```python
#Identify duplicates
df.duplicated()
```

```python
#display duplicate rows
df[df.duplicated()==True]
```

```python
#drop duplicates keeping the last occurence
df.drop_duplicates(keep="last")
```

```python
#drop duplicates removing all duplicates
df.drop_duplicates(keep=False)
```

```python
# drop duplicates based on specific columns
df.drop_duplicates(subset="Age")
```

```python
# drop duplicates based on specific columns
df.drop_duplicates(subset="City")
```

## 2) 5-Fold Cross-Validation

```python
from sklearn.model_selection import KFold
```

```python
df3 = pd.read_csv("btissue.csv")
df3
```

```python
kf = KFold(n_splits=5,shuffle=True,random_state=123)
for fold, (trainIndex,testIndex) in enumerate(kf.split(df3)):
    print(f"Fold Number {fold}")
    print(f"Train indices: {trainIndex}")
    print(f"Test indices: {testIndex}")
    
```

```python
for trainIndex, testIndex in kf.split(df3):
    df3Train = df3.iloc[trainIndex]
    df3Test = df3.iloc[testIndex]
```
