# PCA

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
#LOAD
iris = load_iris()
X=iris.data
```

```python
#APPLY
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print(X.shape)
print(X_pca.shape)
```

```python
df = pd.read_csv("btissue.csv")
df.head()
```

```python
df_pca = pca.fit_transform(df)
print(df.shape)
print(df_pca.shape)
```

# To avoid the above error

```python
df2 = pd.get_dummies(df,drop_first=True)#conversion of nominal to ordinal
```

```python
df2
```

```python
df2_pca = pca.fit_transform(df2)
print(df2.shape)
print(df2_pca.shape)
```

# LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X,y = iris.data,iris.target
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X,y)
print(X.shape)
print(X_lda.shape)
```

# SVD

```python
import numpy as np
from sklearn.decomposition import TruncatedSVD
```

```python
X = np.random.rand(5,4)
```

```python
svd = TruncatedSVD(n_components=2)
X_svd = svd.fit_transform(X)
print(X.shape)
print(X_svd.shape)
```

# Feature Construction

```python
import pandas as pd
data={"length":[20,25,30,32],
     "breadth":[15,23,90,56],
     "price":[20000,30000,40000,50000]}
df = pd.DataFrame(data)
df
```

```python
#Feature construction
df["Area"]=df["length"]*df["breadth"]
df
```
