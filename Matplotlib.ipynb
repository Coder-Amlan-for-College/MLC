```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.linspace(-150,150,100)
y = x**2

plt.plot(x,y)
plt.title("Parabola")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
date=["25/12","26/12","27/12"]
temp = [8.5,10.5,6.8]
plt.plot(date,temp)
plt.title("Date wise temprature")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.grid(True)
plt.yticks(temp)
plt.show()
```
